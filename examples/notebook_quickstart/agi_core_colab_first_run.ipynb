# AGILAB API first run in Colab

This notebook shows the smallest public AGILAB API path from Google Colab.
It installs the published package, resolves the packaged built-in apps, and runs
`mycode_project` locally through `AgiEnv` and `AGI.run(...)`.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/ThalesGroup/agilab/blob/main/examples/notebook_quickstart/agi_core_colab_first_run.ipynb)

```python
import pathlib
from io import UnsupportedOperation as IOUnsupportedOperation

if not hasattr(pathlib, "UnsupportedOperation"):
    pathlib.UnsupportedOperation = IOUnsupportedOperation

from pathlib import Path

import agilab
from agi_cluster.agi_distributor import AGI
from agi_env import AgiEnv

PKG_ROOT = Path(agilab.__file__).resolve().parent
APPS_PATH = PKG_ROOT / "apps" / "builtin"
APP = "mycode_project"
LOG_ROOT = Path.home() / "log" / "execute" / "mycode"

print("Package root:", PKG_ROOT)
print("Apps path:", APPS_PATH)
print("App:", APP)
print("Log root:", LOG_ROOT)
```

```python
app_env = AgiEnv(apps_path=APPS_PATH, app=APP, verbose=0)
result = await AGI.run(
    app_env,
    scheduler="127.0.0.1",
    workers={"127.0.0.1": 1},
    mode=0,
)
result
```

```python
if LOG_ROOT.exists():
    sorted(path.name for path in LOG_ROOT.iterdir())[:20]
else:
    "Run the previous cell first."
```

## What this proves

- `agilab` can be installed directly from PyPI in a notebook runtime.
- The packaged built-in apps are available through the installed wheel.
- `AgiEnv` plus `AGI.run(...)` is enough for a first local API launch.
