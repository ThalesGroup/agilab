# AGILAB API first run in Colab

This notebook shows the smallest public AGILAB API path from Google Colab.
It installs AGILAB from GitHub `main` without a full checkout and runs the built-in
`mycode_project` locally through `AgiEnv` and `AGI.run(...)`.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/ThalesGroup/agilab/blob/main/examples/notebook_quickstart/agi_core_colab_first_run.ipynb)


```python
# Smallest public install path: install AGILAB directly from GitHub main.
# The next cell removes stale /content/agilab source imports if they exist.
!python -m pip uninstall -y agilab agi-cluster agi-node agi-env >/dev/null 2>&1 || true
!python -m pip install -q uv "agilab @ git+https://github.com/ThalesGroup/agilab.git@main"
```

```python
import importlib
import sys
from pathlib import Path

for bad_prefix in ("/content/agilab/src", "/content/agilab"):
    sys.path = [p for p in sys.path if p != bad_prefix and not p.startswith(bad_prefix + "/")]
for name in list(sys.modules):
    if name == "agilab" or name.startswith(("agilab.", "agi_env", "agi_env.", "agi_node", "agi_node.", "agi_cluster", "agi_cluster.")):
        del sys.modules[name]
importlib.invalidate_caches()

from agilab.notebook_colab_support import bootstrap_installed_colab, install_if_needed

ctx = bootstrap_installed_colab()
AGI = ctx.AGI
AgiEnv = ctx.AgiEnv
APPS_PATH = ctx.apps_path
APP = "mycode_project"
LOG_ROOT = Path.home() / "log" / "execute" / "mycode"

print("Installed apps path:", APPS_PATH)
print("Builtin root:", ctx.builtin_root)
print("App:", APP)
print("Log root:", LOG_ROOT)
```

```python
app_env = AgiEnv(apps_path=APPS_PATH, app=APP, verbose=0)
await install_if_needed(AGI, app_env, scheduler="127.0.0.1", workers={"127.0.0.1": 1}, modes_enabled=0)
result = await AGI.run(
    app_env,
    scheduler="127.0.0.1",
    workers={"127.0.0.1": 1},
    mode=0,
)
result
```
