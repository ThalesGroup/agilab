# 03 Compare Predictions

This notebook reads the exported analysis artifacts.
It is the bridge to an AGILAB ANALYSIS page because it already consumes files instead of hidden notebook state.


```python
from pathlib import Path
import json

import pandas as pd

artifact_dir = Path("../analysis_artifacts")
metrics = json.loads((artifact_dir / "forecast_metrics.json").read_text(encoding="utf-8"))
predictions = pd.read_csv(artifact_dir / "forecast_predictions.csv", parse_dates=["date"])
metrics
```

```python
predictions = predictions.set_index("date")
predictions[["y_true", "y_pred"]].plot(title="Observed vs predicted")
```
