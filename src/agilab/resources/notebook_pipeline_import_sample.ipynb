## Build summary artifacts

The importer keeps this markdown as context for the following pipeline stage.

```python
from pathlib import Path
import json
import pandas as pd

Path("artifacts").mkdir(exist_ok=True)
flights = pd.read_csv("data/flights.csv")
summary = {"rows": len(flights), "mean_speed": float(flights["speed"].mean())}
Path("artifacts/summary.json").write_text(json.dumps(summary, indent=2), encoding="utf-8")
summary
```
