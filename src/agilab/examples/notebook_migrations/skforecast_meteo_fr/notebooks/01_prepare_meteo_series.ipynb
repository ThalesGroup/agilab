# 01 Prepare Meteo Series

This notebook is the lightweight source workflow before migration to AGILAB.
It loads a small local daily weather sample shaped like a Meteo-France export.


```python
from pathlib import Path

import pandas as pd

csv_path = Path("../data/meteo_fr_daily_sample.csv")
df = pd.read_csv(csv_path, parse_dates=["date"])
df = df.sort_values("date").set_index("date")
df.head()
```

```python
series = df[["tmax_c"]].copy()
series.rename(columns={"tmax_c": "y"}, inplace=True)
series.plot(title="Daily max temperature")
```

## Migration note

In AGILAB this becomes the `load_clean` step, and the cleaned series is no longer hidden in notebook state.
