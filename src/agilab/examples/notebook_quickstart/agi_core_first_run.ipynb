# AGI-core first local run

This notebook is the smallest code-first AGILab path. It runs `mycode_project`
locally through `AgiEnv` and `AGI.run(...)` without opening the AGILAB web UI.

Before you run it, bootstrap the repository from the AGILab checkout root:

```bash
./install.sh --install-apps --test-apps
uv run --with jupyterlab jupyter lab src/agilab/examples/notebook_quickstart/agi_core_first_run.ipynb
```


````python
from agi_cluster.agi_distributor import AGI
from agilab.notebook_demo import (
    install_if_needed,
    notebook_app_env,
    notebook_local_request,
    notebook_log_root,
)

APP = "mycode_project"  # built-in MyCode example app
app_env = notebook_app_env(APP, verbose=1)
request = notebook_local_request()
print("App:", app_env.app)
print("Log root:", notebook_log_root(app_env))
````

````python
await install_if_needed(app_env, request=request)
result = await AGI.run(app_env, request=request)
result
````

````python
log_root = notebook_log_root(app_env)
print(log_root)
sorted(path.name for path in log_root.iterdir())[:20] if log_root.exists() else "Run the previous cell first."
````

## What to do next

- Open the AGILab web UI once this notebook run succeeds.
