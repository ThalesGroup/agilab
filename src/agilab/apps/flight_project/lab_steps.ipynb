```
from scipy.signal import savgol_filter

column = 'long'
if column not in df.columns:
    raise KeyError(f"Column '{column}' not found in dataframe")

window_length = 7  # must be odd and >= 5
polyorder = 2
if window_length >= len(df):
    window_length = len(df) - 1 if len(df) % 2 == 0 else len(df)
    window_length = max(window_length, 5)
    if window_length % 2 == 0:
        window_length -= 1

df[f"{column}_smooth_gpt_oss"] = savgol_filter(
    df[column].to_numpy(),
    window_length=window_length,
    polyorder=polyorder,
    mode='interp',
)
```

```
import numpy as np

if 'long' not in df.columns:
    raise KeyError("Column 'long' not found in dataframe")

# Lightweight Savitzky-Golay-style smoothing with a fixed 5-point window
window = 5
weights = np.array([0.4, 0.6, 0.8, 0.6, 0.4])
weights /= weights.sum()

series = df['long'].to_numpy(dtype=float)
padded = np.pad(series, (window // 2,), mode='edge')
smoothed = np.convolve(padded, weights, mode='valid')

df['smoothed_long_mistral'] = smoothed
```
