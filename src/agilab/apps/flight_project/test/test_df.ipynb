```python
# Find all parquet files in the given directory.
parquet_files = list(Path("~/data/flight/dataframes").expanduser().glob("*.parquet"))

# Read and concatenate the parquet files.
df = pd.concat((pd.read_parquet(f) for f in parquet_files), ignore_index=True)

# If there is an extra "index" column added by Polars, drop it.
if "index" in df.columns:
    df.drop(columns=["index"], inplace=True)

# Convert the "date" column to datetime, if not already.
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Set the "date" column as the index.
df.set_index("date", inplace=True)
```
